```python
with open('messy_blog.txt', 'w') as f:
    f.write(messy_text)
```

```python
with open('messy_blog.txt', 'r') as file:
    blog_data = file.read()
```

```python
print("clear blog ")
print(blog_data[:100]) 
```

```python
cleaned_data = blog_data.strip()
```

```python
cleaned_data = cleaned_data.replace("###","")
cleaned_data = cleaned_data.replace("@@@","")
cleaned_data = cleaned_data.replace("!!!","")
```

```python
cleaned_data = cleaned_data.lower()
```

```python
words = cleaned_data.split()
```

```python
cleaned_data = " ".join(words)
```

```python
print(cleaned_data)
```

```python
cleaned_data = cleaned_data.replace("gr8","great")
clenead_data = cleaned_data.replace("btw" , "by the way")
```

```python
clenead_data = clenead_data.capitalize()
```

```python
print(clenead_data)
```

```python
with open("clean_blog.txt", "w") as f:
    f.write(clenead_data)
```

```python
print(clenead_data)
```

```python
formatted_text = clenead_data.replace(". ", ".\n")
```

```python
with open("clean_blog.txt", "w") as f:
    f.write(formatted_text)
```
